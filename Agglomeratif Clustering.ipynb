```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Kordinat range 1-100 dengan jumlah 30 titik
#DATASET YANG SAMA DENGAN K-MEANS CLUSTERING SEBELUMNYA
dataset = pd.DataFrame({
    'x':[12,24,54,58,21,88,99,80,92,30,60,71,77,8,9,10,65,2,66,34,35,35,39,70,36,52,37,90,9,1],
    'y':[77,65,42,10,4,10,99,7,64,8,9,18,28,74,59,30,50,89,11,16,28,33,44,54,39,20,88,75,72,30]
})
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dendogram = sch.dendrogram(sch.linkage(dataset, method='ward'))
```

```python
plt.scatter(dataset['x'],dataset['y'])
```

```python
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
```

```python
y_hc = hc.fit_predict(dataset)
```

```python
dataset = np.array(dataset)
plt.scatter(dataset[y_hc == 0,0],dataset[y_hc==0,1], s = 100, c='red')
plt.scatter(dataset[y_hc == 1,0],dataset[y_hc==1,1], s = 100, c='green')
plt.scatter(dataset[y_hc == 2,0],dataset[y_hc==2,1], s = 100, c='blue')
plt.scatter(dataset[y_hc == 3,0],dataset[y_hc==3,1], s = 100, c='yellow')
```
